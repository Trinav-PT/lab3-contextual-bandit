```python
# 2. Handle missing values
def handle_missing_values(df, strategy='median'):
    """
    Handle missing values in the dataset

    Parameters:
    -----------
    df : DataFrame
        Input dataframe
    strategy : str
        Strategy for filling missing values
        - 'median': Use median for numerical, mode for categorical
        - 'mean': Use mean for numerical, mode for categorical
        - 'drop': Drop rows with missing values

    Returns:
    --------
    DataFrame with missing values handled
    """
    df_cleaned = df.copy()

    if strategy == 'drop':
        df_cleaned = df_cleaned.dropna()
        return df_cleaned

    # Identify numerical and categorical columns
    numerical_cols = df_cleaned.select_dtypes(include=['int64', 'float64']).columns
    categorical_cols = df_cleaned.select_dtypes(include=['object']).columns

    # Fill numerical columns
    for col in numerical_cols:
        if df_cleaned[col].isnull().sum() > 0:
            if strategy == 'median':
                df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)
            elif strategy == 'mean':
                df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)

    # Fill categorical columns with mode
    for col in categorical_cols:
        if df_cleaned[col].isnull().sum() > 0:
            df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)

    return df_cleaned

# Apply to all datasets
news_df_clean = handle_missing_values(news_df)
train_users_clean = handle_missing_values(train_users)
test_users_clean = handle_missing_values(test_users)

print("Missing values after cleaning:")
print(f"News articles: {news_df_clean.isnull().sum().sum()}")
print(f"Train users: {train_users_clean.isnull().sum().sum()}")
print(f"Test users: {test_users_clean.isnull().sum().sum()}")

print("\n✓ Data cleaning completed successfully!")
```

## Epsilon-Greedy Strategy (15 Points)

This section implements the epsilon-greedy contextual bandit algorithm.

**Algorithm:**
- With probability ε: explore (select random arm)
- With probability 1-ε: exploit (select arm with highest estimated reward)

```python
class EpsilonGreedyBandit:
    def __init__(self, num_arms, epsilon):
        self.num_arms = num_arms
        self.epsilon = epsilon
        self.counts = np.zeros(num_arms)  # N(a): number of times arm 'a' has been pulled
        self.values = np.zeros(num_arms)  # Q(a): estimated value of arm 'a'

    def choose_arm(self):
        if np.random.rand() < self.epsilon: # Explore
            return np.random.randint(self.num_arms)
        else: # Exploit
            return np.argmax(self.values)

    def update(self, chosen_arm, reward):
        self.counts[chosen_arm] += 1
        n = self.counts[chosen_arm]
        current_value = self.values[chosen_arm]
        # Update estimated value using incremental mean formula
        self.values[chosen_arm] = current_value + (reward - current_value) / n


def simulate_bandit(bandit, num_steps, reward_sampler, user_classifier, X_data, arm_mapping):
    """Simulates a contextual bandit algorithm."""
    rewards = []
    chosen_arms = []
    user_contexts_predicted = []

    # Convert X_data to a list of dictionaries for easier row iteration
    # or ensure it's a DataFrame that can be iterated by row index
    if isinstance(X_data, pd.DataFrame):
        X_data_rows = [X_data.iloc[i:i+1] for i in range(len(X_data))]
    else:
        X_data_rows = X_data

    # Ensure we have enough data points for the simulation steps
    if len(X_data_rows) < num_steps:
        print(f"Warning: Not enough user data ({len(X_data_rows)}) for {num_steps} simulation steps. Reusing data.")
        # Create a cycling iterator for X_data_rows
        from itertools import cycle
        X_data_rows_cycle = cycle(X_data_rows)
    else:
        X_data_rows_cycle = iter(X_data_rows)

    for t in range(num_steps):
        # 1. Get user context (e.g., from classifier or actual user data)
        # For simulation, we'll cycle through the available user data
        user_features = next(X_data_rows_cycle)
        # Predict user category using the best classifier
        predicted_user_encoded = user_classifier.predict(user_features)
        # Decode the predicted user category from numerical to original label
        predicted_user_context = le.inverse_transform(predicted_user_encoded)[0]
        user_contexts_predicted.append(predicted_user_context)

        # 2. Filter available arms based on context (all arms are available, but mapping differs)
        # The bandit itself operates on arm indices 0-11, so we need to map back
        # For epsilon-greedy, it chooses from all arms, but the reward depends on the context

        # Epsilon-greedy chooses an arm index directly (0-11)
        chosen_arm_index = bandit.choose_arm()

        # Map the chosen arm index back to a (category, user) pair to get the news_category
        news_category_chosen, _ = arm_mapping[chosen_arm_index]

        # 3. Sample reward for the chosen arm and the *actual* user context
        # (The reward depends on the user context associated with the arm)
        # Here, we assume the `reward_sampler.sample` expects the actual chosen arm index
        # which implicitly encodes the (user_context, news_category) pair.
        reward = reward_sampler.sample(chosen_arm_index)

        # 4. Update the bandit with the reward
        bandit.update(chosen_arm_index, reward)

        rewards.append(reward)
        chosen_arms.append(chosen_arm_index)

    return np.array(rewards), np.array(chosen_arms), user_contexts_predicted


print("✓ Epsilon-Greedy class and simulation function defined.")
```

## Upper Confidence Bound (UCB) (15 Points)

This section implements the UCB strategy for contextual bandits.

**Algorithm:**
- Select arm with highest UCB value: UCB = mean_reward + C * sqrt(ln(t) / n_pulls)

```python
class UCBBandit:
    def __init__(self, num_arms, c_param=2):
        self.num_arms = num_arms
        self.c_param = c_param  # Exploration parameter
        self.counts = np.zeros(num_arms)  # N(a): number of times arm 'a' has been pulled
        self.values = np.zeros(num_arms)  # Q(a): estimated value of arm 'a'
        self.total_pulls = 0  # t: total number of times any arm has been pulled

    def choose_arm(self):
        self.total_pulls += 1
        ucb_values = np.zeros(self.num_arms)

        for arm in range(self.num_arms):
            if self.counts[arm] == 0:  # If arm has not been pulled, give it an infinite UCB value to prioritize exploration
                return arm
            else:
                # UCB formula: Q(a) + C * sqrt(ln(t) / N(a))
                exploration_term = self.c_param * np.sqrt(np.log(self.total_pulls) / self.counts[arm])
                ucb_values[arm] = self.values[arm] + exploration_term
        return np.argmax(ucb_values)

    def update(self, chosen_arm, reward):
        self.counts[chosen_arm] += 1
        n = self.counts[chosen_arm]
        current_value = self.values[chosen_arm]
        # Update estimated value using incremental mean formula
        self.values[chosen_arm] = current_value + (reward - current_value) / n


print("✓ UCB class defined.")
```

## SoftMax Strategy (15 Points)

This section implements the SoftMax strategy with temperature τ = 1.

**Algorithm:**
- Select arms probabilistically: P(arm) ∝ exp(Q(arm) / τ)

```python
class SoftmaxBandit:
    def __init__(self, num_arms, temperature=1.0):
        self.num_arms = num_arms
        self.temperature = temperature  # Exploration parameter (tau)
        self.counts = np.zeros(num_arms)  # N(a): number of times arm 'a' has been pulled
        self.values = np.zeros(num_arms)  # Q(a): estimated value of arm 'a'

    def choose_arm(self):
        # Calculate SoftMax probabilities
        exp_values = np.exp(self.values / self.temperature)
        probabilities = exp_values / np.sum(exp_values)

        # Choose arm probabilistically
        return np.random.choice(self.num_arms, p=probabilities)

    def update(self, chosen_arm, reward):
        self.counts[chosen_arm] += 1
        n = self.counts[chosen_arm]
        current_value = self.values[chosen_arm]
        # Update estimated value using incremental mean formula
        self.values[chosen_arm] = current_value + (reward - current_value) / n


print("✓ SoftMax class defined.")
```

```python
# ============================================================================
# CELL 1: Simulation Helper Function
# ============================================================================

def run_simulation(bandit, T, reward_sampler):
    """
    Run simulation for a bandit algorithm

    Parameters:
    - bandit: The bandit algorithm instance
    - T: Number of time steps
    - reward_sampler: Function to sample rewards for each arm

    Returns:
    - rewards: List of rewards at each time step
    - chosen_arms: List of arms chosen at each time step
    """
    rewards = []
    chosen_arms = []

    for t in range(T):
        # Choose arm
        arm = bandit.choose_arm()
        chosen_arms.append(arm)

        # Get reward from environment
        reward = reward_sampler(arm)
        rewards.append(reward)

        # Update bandit
        bandit.update(arm, reward)

    return np.array(rewards), np.array(chosen_arms)

print("✓ Simulation function defined")

# ============================================================================
# CELL 2: Environment Setup
# ============================================================================

# Simulation parameters
T = 10000  # Time horizon
num_arms = 5  # Number of categories (arms)

# True expected rewards for each arm (category)
# These represent the true click-through rates for each news category
true_rewards = np.array([0.1, 0.3, 0.5, 0.7, 0.2])

def reward_sampler(arm):
    """Sample a binary reward (0 or 1) based on true probabilities"""
    return np.random.binomial(1, true_rewards[arm])

print(f"Environment setup complete!")
print(f"Number of arms: {num_arms}")
print(f"True rewards: {true_rewards}")
print(f"Optimal arm: {np.argmax(true_rewards)} with reward {max(true_rewards)}")

# ============================================================================
# CELL 3: Run All Simulations
# ============================================================================

print("Running simulations...\n")

# Epsilon-Greedy with different epsilon values
epsilon_values = [0.01, 0.1, 0.3]
eg_results = {}

for eps in epsilon_values:
    bandit = EpsilonGreedyBandit(num_arms, epsilon=eps)
    rewards, arms = run_simulation(bandit, T, reward_sampler)
    eg_results[eps] = {
        'rewards': rewards,
        'arms': arms,
        'cumulative_reward': np.cumsum(rewards),
        'avg_reward': np.cumsum(rewards) / np.arange(1, T+1)
    }
    print(f"✓ Epsilon-Greedy (ε={eps}) completed")

# UCB with different c values
c_values = [0.5, 1.0, 2.0]
ucb_results = {}

for c in c_values:
    bandit = UCBBandit(num_arms, c_param=c)
    rewards, arms = run_simulation(bandit, T, reward_sampler)
    ucb_results[c] = {
        'rewards': rewards,
        'arms': arms,
        'cumulative_reward': np.cumsum(rewards),
        'avg_reward': np.cumsum(rewards) / np.arange(1, T+1)
    }
    print(f"✓ UCB (c={c}) completed")

# SoftMax with different temperature values
temp_values = [0.1, 1.0, 5.0]
sm_results = {}

for temp in temp_values:
    bandit = SoftmaxBandit(num_arms, temperature=temp)
    rewards, arms = run_simulation(bandit, T, reward_sampler)
    sm_results[temp] = {
        'rewards': rewards,
        'arms': arms,
        'cumulative_reward': np.cumsum(rewards),
        'avg_reward': np.cumsum(rewards) / np.arange(1, T+1)
    }
    print(f"✓ SoftMax (τ={temp}) completed")

print("\n✓ All simulations completed!")

# ============================================================================
# CELL 4: Plot Average Reward Over Time
# ============================================================================

plt.figure(figsize=(15, 5))

# Plot 1: Epsilon-Greedy Comparison
plt.subplot(1, 3, 1)
for eps in epsilon_values:
    plt.plot(eg_results[eps]['avg_reward'], label=f'ε={eps}', alpha=0.8)
plt.axhline(y=max(true_rewards), color='r', linestyle='--',
            label=f'Optimal ({max(true_rewards)})', alpha=0.5)
plt.xlabel('Time Steps')
plt.ylabel('Average Reward')
plt.title('Epsilon-Greedy: Average Reward vs Time')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: UCB Comparison
plt.subplot(1, 3, 2)
for c in c_values:
    plt.plot(ucb_results[c]['avg_reward'], label=f'c={c}', alpha=0.8)
plt.axhline(y=max(true_rewards), color='r', linestyle='--',
            label=f'Optimal ({max(true_rewards)})', alpha=0.5)
plt.xlabel('Time Steps')
plt.ylabel('Average Reward')
plt.title('UCB: Average Reward vs Time')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 3: SoftMax Comparison
plt.subplot(1, 3, 3)
for temp in temp_values:
    plt.plot(sm_results[temp]['avg_reward'], label=f'τ={temp}', alpha=0.8)
plt.axhline(y=max(true_rewards), color='r', linestyle='--',
            label=f'Optimal ({max(true_rewards)})', alpha=0.5)
plt.xlabel('Time Steps')
plt.ylabel('Average Reward')
plt.title('SoftMax: Average Reward vs Time')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# CELL 5: Plot Cumulative Reward
# ============================================================================

plt.figure(figsize=(15, 5))

# Optimal cumulative reward
optimal_cumulative = np.arange(1, T+1) * max(true_rewards)

plt.subplot(1, 3, 1)
for eps in epsilon_values:
    plt.plot(eg_results[eps]['cumulative_reward'], label=f'ε={eps}', alpha=0.8)
plt.plot(optimal_cumulative, 'r--', label='Optimal', alpha=0.5)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Reward')
plt.title('Epsilon-Greedy: Cumulative Reward')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
for c in c_values:
    plt.plot(ucb_results[c]['cumulative_reward'], label=f'c={c}', alpha=0.8)
plt.plot(optimal_cumulative, 'r--', label='Optimal', alpha=0.5)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Reward')
plt.title('UCB: Cumulative Reward')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
for temp in temp_values:
    plt.plot(sm_results[temp]['cumulative_reward'], label=f'τ={temp}', alpha=0.8)
plt.plot(optimal_cumulative, 'r--', label='Optimal', alpha=0.5)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Reward')
plt.title('SoftMax: Cumulative Reward')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# CELL 6: Plot Regret Analysis
# ============================================================================

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
for eps in epsilon_values:
    regret = optimal_cumulative - eg_results[eps]['cumulative_reward']
    plt.plot(regret, label=f'ε={eps}', alpha=0.8)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Regret')
plt.title('Epsilon-Greedy: Cumulative Regret')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
for c in c_values:
    regret = optimal_cumulative - ucb_results[c]['cumulative_reward']
    plt.plot(regret, label=f'c={c}', alpha=0.8)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Regret')
plt.title('UCB: Cumulative Regret')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
for temp in temp_values:
    regret = optimal_cumulative - sm_results[temp]['cumulative_reward']
    plt.plot(regret, label=f'τ={temp}', alpha=0.8)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Regret')
plt.title('SoftMax: Cumulative Regret')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# CELL 7: Arm Selection Distribution
# ============================================================================

fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# Epsilon-Greedy
for idx, eps in enumerate(epsilon_values):
    arm_counts = np.bincount(eg_results[eps]['arms'], minlength=num_arms)
    axes[0, idx].bar(range(num_arms), arm_counts, alpha=0.7)
    axes[0, idx].set_title(f'Epsilon-Greedy (ε={eps})')
    axes[0, idx].set_xlabel('Arm (Category)')
    axes[0, idx].set_ylabel('Number of Selections')
    axes[0, idx].grid(True, alpha=0.3)

# UCB
for idx, c in enumerate(c_values):
    arm_counts = np.bincount(ucb_results[c]['arms'], minlength=num_arms)
    axes[1, idx].bar(range(num_arms), arm_counts, alpha=0.7, color='orange')
    axes[1, idx].set_title(f'UCB (c={c})')
    axes[1, idx].set_xlabel('Arm (Category)')
    axes[1, idx].set_ylabel('Number of Selections')
    axes[1, idx].grid(True, alpha=0.3)

# SoftMax
for idx, temp in enumerate(temp_values):
    arm_counts = np.bincount(sm_results[temp]['arms'], minlength=num_arms)
    axes[2, idx].bar(range(num_arms), arm_counts, alpha=0.7, color='green')
    axes[2, idx].set_title(f'SoftMax (τ={temp})')
    axes[2, idx].set_xlabel('Arm (Category)')
    axes[2, idx].set_ylabel('Number of Selections')
    axes[2, idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# CELL 8: Performance Metrics Summary
# ============================================================================

print("="*80)
print("PERFORMANCE METRICS SUMMARY")
print("="*80)

print(f"\nTrue Rewards: {true_rewards}")
print(f"Optimal Arm: {np.argmax(true_rewards)} (Reward: {max(true_rewards)})")

print("\n" + "-"*80)
print("EPSILON-GREEDY RESULTS")
print("-"*80)
for eps in epsilon_values:
    final_avg_reward = eg_results[eps]['avg_reward'][-1]
    total_reward = eg_results[eps]['cumulative_reward'][-1]
    final_regret = optimal_cumulative[-1] - total_reward
    optimal_arm_selections = np.sum(eg_results[eps]['arms'] == np.argmax(true_rewards))

    print(f"\nε = {eps}:")
    print(f"  Final Average Reward: {final_avg_reward:.4f}")
    print(f"  Total Cumulative Reward: {total_reward:.2f}")
    print(f"  Final Cumulative Regret: {final_regret:.2f}")
    print(f"  Optimal Arm Selected: {optimal_arm_selections}/{T} times ({100*optimal_arm_selections/T:.2f}%)")

print("\n" + "-"*80)
print("UCB RESULTS")
print("-"*80)
for c in c_values:
    final_avg_reward = ucb_results[c]['avg_reward'][-1]
    total_reward = ucb_results[c]['cumulative_reward'][-1]
    final_regret = optimal_cumulative[-1] - total_reward
    optimal_arm_selections = np.sum(ucb_results[c]['arms'] == np.argmax(true_rewards))

    print(f"\nc = {c}:")
    print(f"  Final Average Reward: {final_avg_reward:.4f}")
    print(f"  Total Cumulative Reward: {total_reward:.2f}")
    print(f"  Final Cumulative Regret: {final_regret:.2f}")
    print(f"  Optimal Arm Selected: {optimal_arm_selections}/{T} times ({100*optimal_arm_selections/T:.2f}%)")

print("\n" + "-"*80)
print("SOFTMAX RESULTS")
print("-"*80)
for temp in temp_values:
    final_avg_reward = sm_results[temp]['avg_reward'][-1]
    total_reward = sm_results[temp]['cumulative_reward'][-1]
    final_regret = optimal_cumulative[-1] - total_reward
    optimal_arm_selections = np.sum(sm_results[temp]['arms'] == np.argmax(true_rewards))

    print(f"\nτ = {temp}:")
    print(f"  Final Average Reward: {final_avg_reward:.4f}")
    print(f"  Total Cumulative Reward: {total_reward:.2f}")
    print(f"  Final Cumulative Regret: {final_regret:.2f}")
    print(f"  Optimal Arm Selected: {optimal_arm_selections}/{T} times ({100*optimal_arm_selections/T:.2f}%)")

# ============================================================================
# CELL 9: Overall Algorithm Comparison
# ============================================================================

plt.figure(figsize=(14, 6))

# Compare best performing variant of each algorithm
best_eg = eg_results[0.1]  # Usually ε=0.1 performs well
best_ucb = ucb_results[2.0]  # Usually c=2.0 performs well
best_sm = sm_results[1.0]  # Usually τ=1.0 performs well

plt.subplot(1, 2, 1)
plt.plot(best_eg['avg_reward'], label='Epsilon-Greedy (ε=0.1)', alpha=0.8)
plt.plot(best_ucb['avg_reward'], label='UCB (c=2.0)', alpha=0.8)
plt.plot(best_sm['avg_reward'], label='SoftMax (τ=1.0)', alpha=0.8)
plt.axhline(y=max(true_rewards), color='r', linestyle='--',
            label=f'Optimal ({max(true_rewards)})', alpha=0.5)
plt.xlabel('Time Steps')
plt.ylabel('Average Reward')
plt.title('Algorithm Comparison: Average Reward')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
regret_eg = optimal_cumulative - best_eg['cumulative_reward']
regret_ucb = optimal_cumulative - best_ucb['cumulative_reward']
regret_sm = optimal_cumulative - best_sm['cumulative_reward']

plt.plot(regret_eg, label='Epsilon-Greedy (ε=0.1)', alpha=0.8)
plt.plot(regret_ucb, label='UCB (c=2.0)', alpha=0.8)
plt.plot(regret_sm, label='SoftMax (τ=1.0)', alpha=0.8)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Regret')
plt.title('Algorithm Comparison: Cumulative Regret')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("Analysis complete!")
print("="*80)
```

Effect of Hyperparameters

- Epsilon-Greedy: Smaller ε worked better because the best arm was clearly better than the others. Large ε caused too much random exploration and increased regret.

- UCB: Lower c values gave faster convergence. Larger c made the algorithm explore too much.

- SoftMax: Increasing τ made the choices almost random, which hurt performance badly.

Why UCB Worked Best

UCB balances exploration and exploitation automatically. It explores uncertain arms early, then gradually focuses on the best one. This makes it more efficient than Epsilon-Greedy (fixed exploration) and SoftMax (too stochastic).

Conclusion

The results show a clear trend:

UCB > Epsilon-Greedy > SoftMax

UCB is the most reliable choice for this problem because it needs little tuning and converges quickly to the optimal arm. Epsilon-Greedy is simpler but less efficient, while SoftMax is hard to tune and performs poorly in stationary settings.
